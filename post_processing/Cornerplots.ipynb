```python
from cornerplots import cornerPlot, getSamples
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"
```

```python
model = 'kilonova' # kilonova, kilonova_uvboost, shock
datas = 'kilonova'
delay = '0h'
distance = '40' #['160','100','40']
#band = ['no_opticalband_NUV_Dband','no_opticalband_NUV_DD2band','rband_NUV_Dband']
band = ['rband_no_uvband','no_opticalband_NUV_Dband','no_opticalband_NUV_DD2band','rband_NUV_Dband']
legend_texts=['r band', '1 UV band', '2 UV bands', 'r band and 1 UV band']
colors=['blue','orange','green','red']

samples,legend_texts = getSamples(model,datas,delay,distance,band, colors, legend_texts=legend_texts)
figure = cornerPlot(model, samples, legend_texts, colors, plot_density=False, plot_datapoints=False, no_fill_contours=False, fill_contours=False, medians_visible=False, quantilelines_visible=False)

plt.show()
figure.savefig(f'plots/cornerplot_{band}_{model}.png',dpi=300,pad_inches=0.3,bbox_inches='tight')
```
